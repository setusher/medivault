```python
import os
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, BatchNormalization, Flatten, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping, Callback
from tensorflow.keras import regularizers, layers

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from PIL import Image, ImageOps

###############################################################################
# 1. DATASET & DIRECTORIES
###############################################################################
img_dir = 'COVID-19_Lung_CT_Scans'
if os.path.exists(img_dir):
    print(f"Dataset directory found: {img_dir}")
else:
    raise FileNotFoundError(f"Directory not found: {img_dir}")

# Inspect how many images are in each subfolder
for dirname, _, filenames in os.walk(img_dir):
    print(f"{len(filenames)} images found in {dirname}")

###############################################################################
# 2. DATA AUGMENTATION & GENERATORS
###############################################################################
IMAGE_SIZE = 224
BATCH_SIZE = 32

# Enhanced augmentation to help model generalize better
data_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    validation_split=0.2
)

train_gen = data_gen.flow_from_directory(
    img_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    color_mode="rgb",
    shuffle=True,
    class_mode="binary",
    subset="training"
)

val_gen = data_gen.flow_from_directory(
    img_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    color_mode="rgb",
    shuffle=True,
    class_mode="binary",
    subset="validation"
)

print("\nClass indices:", train_gen.class_indices)
classes = list(train_gen.class_indices.keys())
print("Classes:", classes)

###############################################################################
# 3. OPTIONAL: CLASS WEIGHTS (if data is imbalanced)
###############################################################################
def compute_class_weights(generator):
    labels_array = generator.classes
    total = len(labels_array)
    class_0_count = np.sum(labels_array == 0)
    class_1_count = np.sum(labels_array == 1)
    weight_for_0 = (1.0 / class_0_count) * (total / 2.0)
    weight_for_1 = (1.0 / class_1_count) * (total / 2.0)
    return {0: weight_for_0, 1: weight_for_1}

class_weight = compute_class_weights(train_gen)
print(f"Class Weights: {class_weight}")

###############################################################################
# 4. COSINE ANNEALING SCHEDULER (Callback)
###############################################################################
import tensorflow.keras.backend as K

class CosineAnnealingScheduler(Callback):
    def __init__(self, T_max, eta_max, eta_min=0, verbose=0):
        super(CosineAnnealingScheduler, self).__init__()
        self.T_max = T_max
        self.eta_max = eta_max
        self.eta_min = eta_min
        self.verbose = verbose

    def on_epoch_begin(self, epoch, logs=None):
        if not hasattr(self.model.optimizer, 'learning_rate'):
            raise ValueError('Optimizer must have a "learning_rate" attribute.')
        lr = self.eta_min + (self.eta_max - self.eta_min) * (1 + math.cos(math.pi * epoch / self.T_max)) / 2
        self.model.optimizer.learning_rate.assign(lr)
        if self.verbose > 0:
            print(f'\nEpoch {epoch + 1}: CosineAnnealingScheduler setting learning rate to {lr}')

    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        logs['lr'] = K.get_value(self.model.optimizer.learning_rate)

###############################################################################
# 5. BUILD & COMPILE MODEL (TRANSFER LEARNING WITH DENSENET201)
###############################################################################
# Path to your manually downloaded DenseNet201 weights file
WEIGHTS_PATH = 'densenet201_weights_tf_dim_ordering_tf_kernels_notop.h5'  # Update this path

def build_model(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)):
    # Initialize DenseNet201 with no pre-loaded weights
    base_model = DenseNet201(weights=None, include_top=False, input_shape=input_shape)
    # Load your manually downloaded weights
    base_model.load_weights(WEIGHTS_PATH)
    base_model.trainable = False  # Freeze the base model initially

    # Custom top layers
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(512, activation='relu', kernel_regularizer=regularizers.L1L2(l1=1e-5, l2=1e-4))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=base_model.input, outputs=outputs)
    return model

model = build_model()
model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

###############################################################################
# 6. TRAINING (FIRST STAGE)
###############################################################################
cosine_scheduler = CosineAnnealingScheduler(T_max=10, eta_max=1e-3, eta_min=1e-6, verbose=1)
early_stop_1 = EarlyStopping(monitor='val_accuracy', patience=5, min_delta=1e-5,
                             verbose=1, restore_best_weights=True)

history_1 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=20,
    class_weight=class_weight,
    callbacks=[cosine_scheduler, early_stop_1]
)

###############################################################################
# 7. FINE-TUNING (SECOND STAGE)
###############################################################################
def fine_tune_model(mdl, fine_tune_at=300):
    for layer in mdl.layers[fine_tune_at:]:
        if not isinstance(layer, layers.BatchNormalization):
            layer.trainable = True
    mdl.compile(
        optimizer=Adam(learning_rate=1e-5),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return mdl

model = fine_tune_model(model, fine_tune_at=300)

checkpoint = ModelCheckpoint(
    'best_model.keras',
    monitor='val_accuracy',
    verbose=1,
    save_best_only=True,
    mode='max'
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.8,
    patience=5,
    verbose=1,
    mode='min',
    min_delta=1e-4,
    cooldown=1,
    min_lr=1e-7
)

early_stop_2 = EarlyStopping(monitor='val_loss', patience=5, min_delta=1e-5,
                             verbose=1, restore_best_weights=True)

history_2 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=20,
    class_weight=class_weight,
    callbacks=[checkpoint, reduce_lr, early_stop_2]
)

###############################################################################
# 8. LOAD BEST MODEL & EVALUATE
###############################################################################
best_model = load_model('best_model.keras')
val_results = best_model.evaluate(val_gen, return_dict=True)
print("\nValidation Set Evaluation:")
print(f"Accuracy: {val_results['accuracy']*100:.2f}%")
print(f"Loss: {val_results['loss']:.4f}")

###############################################################################
# 9. CONFUSION MATRIX & CLASSIFICATION REPORT
###############################################################################
val_gen.reset()
predictions = best_model.predict(val_gen)
pred_labels = np.round(predictions).astype(int).flatten()

true_labels = val_gen.classes
cm = confusion_matrix(true_labels, pred_labels)
acc = accuracy_score(true_labels, pred_labels)
misclass = 1 - acc

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=True)
plt.title(f'Confusion Matrix\nAccuracy={acc:.4f}; Misclass={misclass:.4f}', size=16)
plt.xlabel('Predicted label', size=14)
plt.ylabel('True label', size=14)
plt.xticks([0.5, 1.5], ['COVID-19', 'Non-COVID-19'], rotation=45, ha="right", size=12)
plt.yticks([0.5, 1.5], ['COVID-19', 'Non-COVID-19'], rotation=45, ha="right", size=12)
plt.show()

report = classification_report(true_labels, pred_labels, target_names=['COVID-19', 'Non-COVID-19'])
print(report)

###############################################################################
# 10. INFERENCE ON INDIVIDUAL IMAGES
###############################################################################
def predict_image(model, img_path, img_size=IMAGE_SIZE):
    img = Image.open(img_path).convert('RGB')
    img = ImageOps.fit(img, (img_size, img_size))
    img_array = np.asarray(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    pred_prob = model.predict(img_array)[0][0]
    if pred_prob < 0.5:
        return "COVID-19", 1 - pred_prob
    else:
        return "Non-COVID-19", pred_prob

sample_images = [
    "COVID-19_Lung_CT_Scans/COVID-19/COVID-19_0001.png",
    "COVID-19_Lung_CT_Scans/COVID-19/COVID-19_0005.png",
    "COVID-19_Lung_CT_Scans/COVID-19/COVID-19_0027.png"
]

for img_path in sample_images:
    pred_class, confidence = predict_image(best_model, img_path)
    print(f"Image: {img_path}")
    print(f"Prediction: {pred_class} | Confidence: {confidence:.4f}\n")
    display_img = Image.open(img_path).convert('RGB')
    plt.imshow(display_img)
    plt.title(f"Predicted: {pred_class}, Confidence: {confidence:.4f}")
    plt.axis('off')
    plt.show()

print("\nAll done!")
```
